```python
import numpy as np

a = np.array([1,2,3],dtype='int32')
print(a)
```

```python
b = np.array([[9.0,8.0,7.0],[6.0,5.0,4.0]])
print(b)
```

```python
#Get dimension
print(a.ndim)
print(b.ndim)
```

```python
#Get shape (# of rows x # of columns)
print(a.shape)
print(b.shape)
```

```python
# Get array length (# of elements)
print(a.size)
print(b.size)
```

```python
# Get type
print(a.dtype)
print(b.dtype)
```

```python
# Get size (bytes per element)
print(a.itemsize)
print(b.itemsize)
```

```python
# Get total array size (total bytes)
print(a.size * a.itemsize)
print(a.nbytes)
print(b.size * b.itemsize)
print(b.nbytes)
```
