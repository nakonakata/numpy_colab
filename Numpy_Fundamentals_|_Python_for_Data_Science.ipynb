<h3 style='border-left: 4px solid purple; padding-left:10px'><i>I. What is numpy?</i></h3>
<hr>
<p>- NumPy is the fundamental package for scientific computing in python.<br>
- It is a Python library that provides a multidimential array and other derived objects for fast operation on arrays. Including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transform, basic linear algebra, basic statistical operations, random simulation and much more.<br>
- At the core of the NumPy package, is the ndarray object. This encapsulates n-dimential arrays of homogeneous data types.
</p><br><br>


<h4><b>Why numpy was created?</b></h4>
<p>Numpy is written in C and uses a wrapper so that we can use it in python. <br>
C has speed and Python has easier syntex, so the creator of this library combined both and created numpy for complex mathmatical operations which needs to be fast, also easier syntex can be very helpful.<br>
Numpy is that library, which made python suitable for data science.
</p><br><br>


<h4><b>Numpy Arrays vs Python Sequences(List, Tuple, Set)</b></h4>
<ul>
    <li>Numpy arrays have a fixed size at creation, unline python lists(which can grow dynamically). Changing the size of an numpy array will create a new array and delete the original.</li>
    <li>The elements in a numpy array are all required to be of the dame data type, and thus will be the same size in memory.</li>
    <li>Numpy arrays facilitate advanced mathematical and other types of operations on large numbers of data with greater efficiency and less code.</li>
    <li>Most of the common libraries used for data sciences uses numpy to some extend and often output Numpy arrays.</li>
</ul>

<h3 style='border-left: 4px solid purple; padding-left:10px'><i>II. Creating Numpy Arrays</i></h3>
<hr>

```python
import numpy as np
```

```python
mylist = [0,1,2,3,4]
print(mylist)
print(type(mylist))
```

##### **1D Array (Vector)**

```python
vector = np.array(mylist) #this converts list to numpy array
print(vector)
print(type(vector))
```

##### **2D Array (Matrix)**

```python
matrix = np.array([[1,2],[2,3]])
print(matrix)
```

##### **3D Array (Tensor)**

```python
tensor = np.array([[[1,2],[3,4]],[[1,2],[3,4]]])
print(tensor)
```

##### **numpy array with different data types**

```python
print(np.array(mylist, dtype = float))
print(np.array(mylist, dtype = bool))
print(np.array(mylist, dtype = complex))
```

##### **arange function in numpy works just like range function in python**

```python
print(np.arange(1,10))
print(np.arange(1,10,2))
```

##### **shape function return the shape of an array**

```python
np.arange(1,10).shape
```

##### **reshape function can be used to change the shape of a function to any dimention**

```python
arr = np.arange(1,13) #1D
print("2D: \n",arr.reshape(2,6))#2D
print("3D: \n",arr.reshape(2,2,3))#3D
print("Shape: ",arr.reshape(2,2,3).shape)
```

##### **np.ones, np.zeros, np.random can create a numpy array having all the indices as one, zeros and random number**
> Used to initialize arrays

```python
print('1D: \n',np.ones(6)) #1D
print('2D: \n',np.ones((6,2))) #2D
print('3D: \n',np.ones((2,2,6))) #2D
```

```python
print('1D: \n',np.zeros(6)) #1D
print('2D: \n',np.zeros((6,2))) #2D
print('3D: \n',np.zeros((2,2,6))) #2D
```

```python
print('random(floating 0 to 1): \n',np.random.random((3,4)))
```

##### **np.linspace(linearly spaced) will generate number of points to a given range**
> **np.linspace(from,to,number of items)**

```python
print(np.linspace(-5,5,10))
print(np.linspace(-5,5,11))
print(np.linspace(-5,5,11,dtype=int))
```

##### **np.identity is used to generate an identitity matrix, which takes only one argument**

```python
print(np.identity(5,dtype=int)) #only diagonal items 1, else 0
```

<a id='functions'></a>
<h3 style='border-left: 4px solid purple; padding-left:10px'><i>VI. Array Functions</i></h3>
<hr>

```python
arr = np.random.random((3,3))
arr = np.round(arr*10)
arr = arr.astype(np.int32)
arr
```

##### **max | min | sum | prod**

```python
print("max ==>",np.max(arr))
print("min ==>",np.min(arr))
print("sum ==>",np.sum(arr))
print("prod =>",np.prod(arr))
```

```python
#you can perform row or columnwise using axis
# 0 -> col
# 1 -> row

#max
print("Column Max ==> ",np.max(arr,axis=0))
print("Row Max =====> ",np.max(arr,axis=1))
#min
print("Column Min ==> ",np.min(arr,axis=0))
print("Row Min =====> ",np.min(arr,axis=1))
#sum
print("Column Sum ==> ",np.sum(arr,axis=0))
print("Row Sum =====> ",np.sum(arr,axis=1))
```

##### **Mean | Median | STD | Var**

```python
print("mean ===>",np.mean(arr))
print("median =>",np.median(arr))
print("std ====>",np.std(arr))
print("var ====>",np.var(arr))
```

```python
#max
print("Column mean ==> ",np.mean(arr,axis=0))
print("Row mean =====> ",np.mean(arr,axis=1))
#min
print("Column median > ",np.median(arr,axis=0))
print("Row median ===> ",np.median(arr,axis=1))
#sum
print("Column STD ===> ",np.std(arr,axis=0))
print("Row STF ======> ",np.std(arr,axis=1))
```

##### **Trigonometric Functions**

```python
print("sin:\n",np.sin(arr))
print("cos:\n",np.cos(arr))
```

##### **Dot Product (Matrix Multiplication)**

```python
m = np.arange(12).reshape(3,4)
n = np.arange(12).reshape(4,3)
print("m ===> \n",m)
print("n ===> \n",n)
dotProduct = np.dot(m,n)
print("m.n => \n",dotProduct)
```

##### **log and exponent**

```python
print("log ===>\n",np.log(arr))
print("exp ===>\n",np.exp(arr))
```

```python
x = np.arange(10,dtype=np.int32)
y = np.arange(12,dtype=np.int32).reshape(3,4)
z = np.arange(27).reshape(3,3,3)
```

##### **1D Slicing**

```python
x
```

```python
#everything is same as list in python
x[3:6]
```

```python
x[2:7:2]
```

```python
x[-5:-2]
```

##### **2D Slicing**

```python
y
```

```python
# y[row,column]
print("0th row: 0th column ==>",y[0,0])
print("1st row: 2nd column ==>",y[1,2])
print("2nd row: 3rd column ==>",y[2,3])
```

```python
print("1st row:",y[1,:])
print("0th row:",y[:,0])
print("3rd row:",y[:,3])
```

```python
print("row 1 & 2 : column 0 & 1 ==>\n",y[1:3,0:2])
```

##### **3D slicing**

```python
z
```

```python
print(z[0,0,0])
print(z[1,0,0])
print(z[0,2,2])
print(z[-1,-1,-1])
```

```python
print(z[1 ,1:3 ,1:3 ])
```

```python
print(z[1:3 ,1:3 ,2 ])
```

```python
print(z[2 ,1: ,1: ])
```
